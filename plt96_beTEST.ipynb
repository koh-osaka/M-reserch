```python
y0=list(data[84:85].values[0][2:])
y1=list(data[85:86].values[0][2:])
y2=list(data[86:87].values[0][2:])
y3=list(data[87:88].values[0][2:])
y4=list(data[88:89].values[0][2:])
y5=list(data[89:90].values[0][2:])
```

```python
y0
```

```python
y0=[float(s)-float(6) for s in y0]   #単位をm^3に
y1=[float(s)-float(6) for s in y1]
y2=[float(s)-float(6) for s in y2]
y3=[float(s)-float(6) for s in y3]
y4=[float(s)-float(6) for s in y4]
y5=[float(s)-float(6) for s in y5]
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)
ax.set_xlabel('Tempreture(log)[eV]')
ax.set_ylabel('Ionization rate cofficient[m^3/s]')
ax.plot(x,y0,label="1")
ax.plot(x,y1,label="2")
ax.plot(x,y2,label="3")
ax.plot(x,y3,label="4")
ax.plot(x,y4,label="5")
ax.plot(x,y5,label="6")
ax.set_xlim(0,)
ax.set_ylim(-36,)
ax.tick_params(direction='in' , bottom=True ,top=True ,left=True ,right=True)
ax.set_title("Temperature dependence of coefficients(n=10^17 [m^-3])" ,y=-0.25)
ax.legend(loc=0)
```
